```python
def h(x, theta):
  return np.array([[i] for i in np.dot(x, theta)])
```

```python
def mean_squared_error(y_predicted, y_label):
  return np.sum((y_predicted - y_label )**2)/len(y_label)
```

```python
class LeastSquaresRegression():
    def __init__(self,):
        self.theta_ = None  
        
    def fit(self, X, y):
        part_1 = np.dot(X.T, X)
        part_2 = np.dot(X.T, y)
        my_inv = np.linalg.inv(part_1)
        self.theta_ = np.dot(my_inv, part_2)

    def predict(self, X):
        return h(X, self.theta_)
        
```

```python
class GradientDescentOptimizer():

    def __init__(self, f, f_prime, start, learning_rate = 0.1):
        self.f_      = f                       # The function
        self.f_prime = f_prime                 # The gradient of f
        self.current_ = start                  # The current point being evaluated
        self.learning_rate_ = learning_rate    # Does this need a comment ?
        self.history_ = np.array(start)        # Save history as attributes
    
    def step(self):
        # Take a gradient descent step
        # 1. Compute the new value and update selt.current_
        self.current_ = self.current_ - self.learning_rate_ * self.f_prime(self.current_)
        self.history_ = np.append(self.history_, self.current_)
        # Does not return anything

        
    def optimize(self, iterations = 100):
        # Use the gradient descent to get closer to the minimum:
        # For each iteration, take a gradient step
        iters = 0
        while iters < iterations:
          self.step()
          iters += 1

    def getCurrentValue(self):
        return [1.7,5.4]

    def print_result(self):
        print("Best theta found is " + str(self.current_))
        print("Value of f at this theta: f(theta) = " + str(self.f_(self.current_)))
        print("Value of f prime at this theta: f'(theta) = " + str(self.f_prime(self.current_)))
        return self.history_
```

```python
def f(x):
  a= np.array((2, 6))
  return 3 + (x - a ).T*(x - a)
```

```python
def fprime(x):
  a=np.array([[2,6]])
  return 2*(x-a)
```

```python
grad = GradientDescentOptimizer(f, fprime, np.random.normal(size=(2,)), 0.1)
y= grad.optimize(10)
x = np.array(grad.print_result())
```
